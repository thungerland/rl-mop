```python
import gymnasium as gym
import minigrid
from minigrid.wrappers import *
from minigrid.envs.babyai import *
import matplotlib.pyplot as plt
```

We want to have a quick way to visualise the environments to know what to expect.

```python
# Now you can create the environment
env = gym.make('BabyAI-GoToObj-v0', render_mode='rgb_array')
env = RGBImgObsWrapper(env)
obs = env.reset()
```

```python
plt.imshow(obs[0]['image'])
```

## Go To Red Ball Grey

```python
# Multiple grids 
fig, axes = plt.subplots(3, 4, figsize=(12, 9))
axes = axes.flatten()

for i, ax in enumerate(axes):
    # Create a fresh env instance
    env = gym.make('BabyAI-GoToRedBallGrey-v0', render_mode='rgb_array')
    env = RGBImgObsWrapper(env)

    # Reset with a different seed for each instance
    obs, info = env.reset(seed=i)

    ax.imshow(obs['image'])
    ax.set_title(f"Seed {i}")
    ax.set_xlabel(obs['mission'], fontsize=9)
    ax.set_xticks([])
    ax.set_yticks([])

    env.close()

plt.tight_layout()
plt.gca()
```
